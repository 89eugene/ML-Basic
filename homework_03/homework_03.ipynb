## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
input_value = input('Введите строку: ');


def convert_register(value: str) -> str:
    if is_comal_case(value):
        return convert_to_snake_case(value)
    else:
        return convert_to_camel_case(value)


def is_comal_case(value: str) -> bool:
    return (value != value.lower() and value != value.upper() and "_" not in value)


def convert_to_camel_case(snake_str_value: str) -> str:
    result = ''
    is_next_upper = False
    for _ in range(len(snake_str_value)):
        char = snake_str_value[_]
        if char == '_':
            is_next_upper = True
        else:
            if is_next_upper:
                result += char.upper()
                is_next_upper= False
            else:
                 result += char.lower()

    return result


def convert_to_snake_case(camel_str_value: str) -> str:
    result = ''
    for _ in range(len(camel_str_value)):
        char = camel_str_value[_]
        if _ != 0 and char == char.upper():
            result += '_' + char.lower()
        else:
            result += char.lower()

    return result


print(convert_register(input_value))
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
from datetime import datetime

input_value = input('Введите дату: ')


def is_valid_date(date_string, format="%d.%m.%Y") -> bool:
    try:
        datetime.strptime(date_string, format)
        return True
    except ValueError:
        return False



print(is_valid_date(input_value))
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
input_value = int(input('Введите число: '))



def is_prime(n):

    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True


print(is_prime(input_value))
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
is_break = False
dictionary = {}


def is_valid_age(age: int) -> bool:
    return 18 <= age <= 60

def map_id(value: str) -> str:
    if len(value) < 8:
        result = ''
        for _ in range(8 - len(value)):
            result += '0'

        return result + value

    return value


def check_and_convert_fio(value: str) -> str:
    first_letter = value[0]
    if (first_letter.islower()):
        return value[:1].upper() + value[1:]
    return value


while not is_break:
    input_value = input("'Введите Имя' 'Фамилию' 'Возраст' 'ID': ")
    if (input_value == ''):
        is_break = True

    if not is_break:
        split_input_value = input_value.split(' ')
        name = split_input_value[0]
        surname = split_input_value[1]

        if not name.isalpha() or not surname.isalpha():
            print('Имя или Фамилия содержат не только буквы')
            break

        name = check_and_convert_fio(name)
        surname = check_and_convert_fio(surname)

        age = split_input_value[2]

        if not is_valid_age(int(age)):
            print('Возраст должен быть больше 17 и меньше 61')
            break

        id = map_id(split_input_value[3])

        if dictionary.get(id) != None:
            print('Такой идентификатор уже заведен')
            break
        else:
            dictionary[id] = (name, surname, age)




def print_result(dictionary):
    print(f' ID       Имя   Фамилия   Возраст')
    for key, value in dictionary.items():
        print(f' {key} {value[0]}   {value[1]}   {value[2]}' )



print_result(dictionary)
```
