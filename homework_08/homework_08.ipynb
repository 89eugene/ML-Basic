# Линейная алгебра, математический анализ

```python
import numpy as np
import matplotlib.pyplot as plt


n = 100

np.random.seed(42)
A = np.random.randn(n, n)
```

```python
U, s, Vt = np.linalg.svd(A, full_matrices=True)
S = np.diag(s)


ranks = np.arange(2, n + 1)
errors = []

for r in ranks:

    A_tilde = U[:, :r] @ S[:r, :r] @ Vt[:r, :]

    error = np.linalg.norm(A - A_tilde, 'fro')
    errors.append(error)
```

```python


plt.figure(figsize=(10, 6))
plt.plot(ranks, errors, 'b-', linewidth=2)
plt.xlabel('Ранг аппроксимации (r)')
plt.ylabel('Ошибка аппроксимации $E(r)$')
plt.title('Зависимость ошибки аппроксимации от ранга')
plt.grid(True, alpha=0.3)
plt.xlim(2, n)
plt.tight_layout()
plt.show()
```
